```python
q1=dfship["Age"].quantile(0.25)
```

```python
q1
```

```python
q3
```

```python
iqr=q3-q1
```

```python
lower=q1-(1.5*iqr)
```

```python
upper=q3+(1.5*iqr)
```

```python
lower
```

```python
upper
```

```python
dfship["Age"]<lower
```

```python
dfship["Age"]>upper
```

```python
dfship[(dfship["Age"]<lower) | (dfship["Age"]>upper)]
```

```python
dfship["Age"].map(lambda x: lower if x<lower else upper if x>upper else x)
```
