```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly2=PolynomialFeatures()
```

```python
poly2
```

PolynomialFeatures() is applied ONLY to X, NOT y.

Because polynomial regression works by expanding the input features, not the output.

PolynomialFeatures() is used to create new features like x, x, x from your original X so the model can learn curved relationships.

It expands your input data into polynomial terms.


```python
poly2.fit(X)
```

```python
poly2.transform(X)
```

```python
X_poly2=poly2.transform(X)
```

```python
X_poly2
```

```python
lr2=LinearRegression()
```

```python
lr2
```

```python
lr2.fit(X_poly2,y)
```

```python
lr2.score(X_poly2,y)
```

```python
y_pred1=lr2.predict(X_poly2)
```

```python
y_pred1
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred1)
```
