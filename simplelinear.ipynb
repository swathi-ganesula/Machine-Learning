```python
sc_list=[]
for i in range(500):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=i)
    linreg=LinearRegression()
    linreg.fit(X_train,y_train)
    sc=linreg.score(X_test,y_test)
    sc_list.append(sc)
m=max(sc_list)
print(m)
id=sc_list.index(m)
print(id)
```
